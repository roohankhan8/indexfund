```python
import pandas as pd

# Path to your Excel file
input_file = "merged_frm_output.xlsx"
output_file = "filtered_file.xlsx"

# List of columns you want to keep (can include partial matches)
columns_to_keep = ["SYMBOL", "COMPANY", "PRICE", "Volume", "VOLUME"]  # Example columns to keep

# Load the Excel file
excel_file = pd.ExcelFile(input_file)

# Dictionary to store the modified sheets
filtered_sheets = {}

# Iterate through each sheet
for sheet_name in excel_file.sheet_names:
    df = excel_file.parse(sheet_name)
    
    # Filter the columns to keep
    filtered_columns = [col for col in df.columns if any(keyword in col for keyword in columns_to_keep)]
    filtered_df = df[filtered_columns]
    
    # Save the filtered dataframe to the dictionary
    filtered_sheets[sheet_name] = filtered_df
    print(f"Processed: {sheet_name}")


# Save all filtered sheets to a new Excel file
with pd.ExcelWriter(output_file) as writer:
    for sheet_name, filtered_df in filtered_sheets.items():
        filtered_df.to_excel(writer, sheet_name=sheet_name, index=False)
        print(f"Added: {sheet_name}")

print(f"Filtered data saved to {output_file}")
```
