1. Walk-Forward Validation (THIS IS NON-NEGOTIABLE)

If you skip or mess this up, your entire project looks naïve.

What problem are we solving?

Financial time series cannot be randomly shuffled. Markets have memory, regimes, and causality flows forward in time only.

Random train-test split = future leaking into the past = invalid research.

Correct logic (explain this clearly in your report)

You train on past data, test on future data, then roll forward.

Example structure:

Train: Jan 2019 → Dec 2020

Test: Jan 2021 → Jun 2021

Then:

Train: Jan 2019 → Jun 2021

Test: Jul 2021 → Dec 2021

Repeat until data ends.

This mimics how a real fund manager would deploy the model.

What exactly are you validating?

You apply walk-forward validation to both models:

Fund Flow Prediction model

Return Prediction / Classification model

Metrics you report

For flow regression:

MAE (mean absolute error)

RMSE

Directional accuracy (did it at least predict inflow vs outflow correctly?)

For classification (later):

Accuracy

Precision / Recall

Confusion Matrix

Examiner translation

“The model was evaluated using walk-forward validation to replicate real-time investment decision making and avoid look-ahead bias.”

That sentence alone signals maturity.
