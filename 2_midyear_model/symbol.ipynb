```python
def plot_event_window(symbol, rebalance_date, window=10):
    start = pd.Timestamp(rebalance_date) - pd.Timedelta(days=window)
    end = pd.Timestamp(rebalance_date) + pd.Timedelta(days=window)

    temp = df[(df["SYMBOL"] == symbol) &
              (df["DATE"] >= start) &
              (df["DATE"] <= end)]

    temp["PRED"] = flow_model.predict(temp[features])

    plt.figure(figsize=(12,5))
    plt.bar(temp["DATE"], temp["FLOW_PROXY"], label="Actual")
    plt.plot(temp["DATE"], temp["PRED"], label="Predicted")
    plt.axvline(pd.Timestamp(rebalance_date), color="red", linestyle="--", label="Rebalance")
    plt.legend()
    plt.title(f"Event Study: Fund Flows around Rebalance for {symbol}")
    plt.show()
```
