```python
import pandas as pd

# Path to your Excel file
input_file = "merged_2024_output.xlsx"
output_file = "merged_2024_nbppgi_companies.xlsx"

# Define your desired symbols to filter by
desired_symbols = ['BAHL','EFERT','ENGRO','FFC','HBL','HUBC','LUCK','MCB','MEBL','MTL']  # Replace with your desired symbols

# Define the row filtering function
def filter_rows_by_symbol(df):
    if "SYMBOL" in df.columns:
        return df[df["SYMBOL"].isin(desired_symbols)]
    return df  # If SYMBOL column is not present, return the original dataframe

# Load the Excel file
excel_file = pd.ExcelFile(input_file)

# Dictionary to store the filtered sheets
filtered_sheets = {}

# Iterate through each sheet
for sheet_name in excel_file.sheet_names:
    df = excel_file.parse(sheet_name)

    # Filter rows based on SYMBOL
    filtered_df = filter_rows_by_symbol(df)

    # Save the filtered dataframe to the dictionary
    filtered_sheets[sheet_name] = filtered_df
    print(f"Processed: {sheet_name}")

# Save all filtered sheets to a new Excel file
with pd.ExcelWriter(output_file) as writer:
    for sheet_name, filtered_df in filtered_sheets.items():
        filtered_df.to_excel(writer, sheet_name=sheet_name, index=False)
        print(f"Added: {sheet_name}")

print(f"Filtered rows by SYMBOL saved to {output_file}")
```
