```python
import os
import pandas as pd

# Folder containing the Excel files
folder_path = "E:/Codes/Python/web scraping/cons/2024data"

# Output file path for the merged file
output_file = "merged_2024_output.xlsx"

# Create a Pandas ExcelWriter object to write all files into a single Excel file
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    # Loop through all files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith(".xls"):  # Process only .xls files
            file_path = os.path.join(folder_path, filename)
            
            try:
                # Read the .xls file into a DataFrame
                df = pd.read_excel(file_path)
                
                # Use the file name (without extension) as the sheet name
                sheet_name = os.path.splitext(filename)[0]
                
                # Write the DataFrame to a new sheet in the merged Excel file
                df.to_excel(writer, sheet_name=sheet_name, index=False)
                
                print(f"Added {filename} to the merged file.")

            except Exception as e:
                print(f"Error processing '{filename}': {e}")

print(f"All files merged into {output_file}")
```
